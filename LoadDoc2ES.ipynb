```python
import pickle 
import base64
import os 
from elasticsearch import Elasticsearch
 
```

```python
cluster  = ['9.110.74.161:9200','9.110.74.115:9200','9.110.74.179:9200']
es = Elasticsearch(cluster,timeout=300)
index_name = 'gov_file_index'
type_name='my_type'
pipeline_name='attachment'
```

```python
with open ('filePaths', 'rb') as fp:
    filePaths = pickle.load(fp)
```

```python
print(os.path.basename(filePaths[39]) + ' Size:'+str(os.path.getsize(filePaths[39])))

1024*1024*100 #100MB
```

```python

#i_start = 1
i_start = 40
count = len(filePaths)
for fi in range(i_start, count):
    i=fi
    fPath = filePaths[fi]
    
    if os.path.getsize(fPath) < 1024*1024*100: #100MB
        with open(fPath, 'rb') as file:
            file_code =  base64.b64encode(file.read()).decode('ascii')
            print(fPath)
            doc={
                'file_name': os.path.basename(fPath),
                'data': file_code,
            }
            res = es.index(index=index_name, 
                           doc_type=type_name, 
                           id=i, 
                           pipeline=pipeline_name, 
                           body=doc, 
                           #refresh ='true'
                          )
```
